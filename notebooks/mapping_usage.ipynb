```python
from pyramidtransform import transform_image
from scanip.visualize import display_points
from utils import load_relations_from_yaml

relations = load_relations_from_yaml('./relations_1')
points_camera_1 = transform_image('./camera_1.jpg', relations)
%store points_camera_1

relations_x = load_relations_from_yaml('./relations_2')
points_camera_2 = transform_image('./camera_2.jpg', relations_x)
%store points_camera_2

%matplotlib inline
display_points(points_camera_1)
display_points(points_camera_2)
display_points(points_camera_1 + points_camera_2)
```

```python
import numpy as np

from scanip.scripts.display_asc import plot_data_with_matplotlib

Z_COORDINATE = 1

all_points = points_camera_1 + points_camera_2

# Z Coordinate from motor step will be attached
point_cloud = np.array([
    [int(point.x) for point in all_points],
    [Z_COORDINATE] * len(all_points),
    [point.y for point in all_points],
])

plot_data_with_matplotlib(point_cloud)
```
